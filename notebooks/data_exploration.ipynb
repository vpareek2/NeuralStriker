```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from IPython.display import display
```

```python
df_2022 = pd.read_csv('../data/final/2022.csv').drop(columns=['Unnamed: 0'])
df_2023 = pd.read_csv('../data/final/2023.csv').drop(columns=['Unnamed: 0'])
```

```python
dataframes = [df_2022, df_2022]
for df in dataframes:
    display(df.head(3))
```

```python
df_2022.describe()
```

```python
df_2023.describe()
```

```python
sns.histplot(df_2022['Gls'], kde=True)
plt.title('Histogram of Goals Scored per 90 minutes in 2022') 
plt.xlabel('Goals')
plt.ylabel('Count')

plt.show()
```

```python
sns.histplot(df_2023['Gls'], kde=True, bins=45)
plt.title('Histogram of Goals Scored per 90 minutes in 2023') 
plt.xlabel('Goals')
plt.ylabel('Count')

plt.show()
```

```python
plt.scatter(df_2023['Gls'], df_2023['xG'])
plt.xlabel('Goals')
plt.ylabel('xG')
plt.title('Goals vs Expected Goals in 2023')
plt.show()
```

```python
plt.scatter(df_2022['Gls'], df_2022['xG'])
plt.xlabel('Goals')
plt.ylabel('xG')
plt.title('Goals vs Expected Goals in 2022')
plt.show()

# Big outliers.
```

```python
numeric_columns2022 = df_2022.select_dtypes(include=[np.number])
numeric_columns2023 = df_2023.select_dtypes(include=[np.number])

correlation_matrix2022 = numeric_columns2022.corr()
correlation_matrix2023 = numeric_columns2023.corr()
```

```python
plt.figure(figsize=(7, 3))  # Can adjust the size as necessary
sns.heatmap(correlation_matrix2022, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap 2022')
plt.show()
```

```python
plt.figure(figsize=(7, 3))  # Can adjust the size as necessary
sns.heatmap(correlation_matrix2023, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap 2023')
plt.show()
```
