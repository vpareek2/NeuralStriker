```python
import pandas as pd
from IPython.display import display
```

```python
df_top5_2022 = pd.read_csv('../../data/processed/top5_2022.csv').drop(columns=['Unnamed: 0'])
df_top5_2023 = pd.read_csv('../../data/processed/top5_2023.csv').drop(columns=['Unnamed: 0'])
df_other_2022 = pd.read_csv('../../data/processed/otherleagues2022.csv').drop(columns=['Unnamed: 0'])
df_other_2023 = pd.read_csv('../../data/processed/otherleagues2023.csv').drop(columns=['Unnamed: 0', 'Unnamed: 0.1', 'Unnamed: 0.2'])
```

```python
dataframes = [df_top5_2022, df_top5_2023, df_other_2022, df_other_2023]
for df in dataframes:
    display(df.head(4))
```
